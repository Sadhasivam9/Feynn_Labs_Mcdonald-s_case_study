```python
# Get the summary of PCA results
explained_variance = MD_pca.explained_variance_
proportion_of_variance = MD_pca.explained_variance_ratio_
cumulative_proportion = np.cumsum(proportion_of_variance)

# Print the summary
print("Importance of components:")
for i in range(len(explained_variance)):
    print(f"PC{i+1} Standard deviation: {np.round(np.sqrt(explained_variance[i]), 4)}")
    print(f"PC{i+1} Proportion of Variance: {np.round(proportion_of_variance[i], 4)}")
    print(f"PC{i+1} Cumulative Proportion: {np.round(cumulative_proportion[i], 4)}")
    print()
```

```python
print("Standard deviations (1, .., p=11):")
print(np.round(MD_pca.singular_values_, 1))
print("Rotation (n x k) = (11 x 11):")
print(np.round(MD_pca.components_, 2))
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Create a PCA object
pca = PCA()

# Fit the PCA model
pca.fit(MD_x_scaled)

# Get the projected data
projected_data = pca.transform(MD_x_scaled)

# Plot the projected data
plt.scatter(projected_data[:, 0], projected_data[:, 1], color='grey')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection")

# Add annotations for each column with arrows
for i, column_name in enumerate(MD_x):
    x = projected_data[i, 0]
    y = projected_data[i, 1]
    plt.annotate(column_name, (x, y), xytext=(5, 5), textcoords="offset points", ha='left', va='bottom')
    plt.arrow(0, 0, x, y, color='red', linestyle='--', linewidth=0.5)

plt.show()
```

```python

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Set the random seed
np.random.seed(1234)

# Standardize the data
scaler = StandardScaler()
MD_x_scaled = scaler.fit_transform(MD_x)

# Perform PCA
pca = PCA()
pca.fit(MD_x_scaled)
projected_data = pca.transform(MD_x_scaled)

# Perform clustering using KMeans
inertias = []
segment_numbers = range(2, 9)
for num_segments in segment_numbers:
    kmeans = KMeans(n_clusters=num_segments, n_init=10, random_state=1234)
    kmeans.fit(projected_data)
    inertias.append(kmeans.inertia_)

# Plot the inertias
plt.bar(segment_numbers, inertias)
plt.xlabel("Number of Segments")
plt.ylabel("Inertia")
plt.title("Model Selection")
plt.show()
```

```python
kmeans = KMeans(4)
kmeans.fit_predict(MD_x_scaled)
pca_df_kmeans = pd.concat([pca_df, pd.DataFrame({'cluster':kmeans.labels_})], axis=1)
```
